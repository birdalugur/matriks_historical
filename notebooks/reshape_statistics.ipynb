```python
import pandas as pd
```

```python
stats = pd.read_csv('statistics.csv',index_col=0)
```

```python
stats=stats.drop(['dof','chi2','oddsratio'])
```

```python
data = stats.reset_index()
```

```python
x=stats.stack().reset_index()
```

```python
x.columns=['test','symbol','value']
```

```python
y=x.sort_values(['test','value'],ascending=False)
```

```python
y=y.reset_index(drop=True)
```

```python
group = y.groupby('test')
```

```python
group_name  = list(group.groups.keys())
```

```python
list_test = []
```

```python
for name in group_name:
    res = group.get_group(name).pivot(index='symbol',columns='test',values='value')\
    .sort_values(by = name).reset_index()
    list_test.append(res)
```

```python
result = pd.concat(list_test,axis=1)
```
